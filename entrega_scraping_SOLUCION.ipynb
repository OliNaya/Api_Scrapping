```python
#Scrapea a pagina de sensacine para obter o nome da pelicula e data de estreno dos estreos mais esperados.
#https://www.sensacine.com/
```

```python
url = 'https://www.sensacine.com/'
paxina = requests.get(url)
soup = BeautifulSoup(paxina.content, 'html.parser')
div=soup.find_all('div', {'class':'title-wrapper'})
```

```python

for nova in div:
    print(nova.find('a').text,nova.find('div',{'class': 'list-entity-item-director'}))
novas = []
for nova in div:
    #print(nova.find('h2').text)
    novas.append([nova.find('a').text,nova.find('div',{'class': 'list-entity-item-director'}).text])
print(novas)
```

```python
#Crear un dataFrame cos datos obtidos 
```

```python
df_titulos = pd.DataFrame(novas,columns = ['título','DataEstreno'])
df_titulos = df_titulos.set_index('DataEstreno')


df_titulos
```

```python
#Obten da paxina gastronomia de galicia o nome de todas as receitas de Marisco e crea con elas un DataFrame
#http://www.gastronomiadegalicia.com/v_portal/apartados/apartado.asp?te=499
```

```python
url = 'http://www.gastronomiadegalicia.com/v_portal/apartados/apartado.asp?te=499'
paxina = requests.get(url)
soup = BeautifulSoup(paxina.content, 'html.parser')
receitas=[]
result = soup.find_all('li')
for rec in result:
   
    solo=rec.find_all('a',{'title':'Ver'})
    for s in solo:
        receitas.append(s.text.strip('(\r)(\n)'))
        
import pandas as pd
rec = pd.DataFrame(receitas,columns = ['Receita'])  
rec
```
